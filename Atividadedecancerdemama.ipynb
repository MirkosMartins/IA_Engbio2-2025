```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Carregando o dataset
cancer = load_breast_cancer()
X = cancer.data
y = cancer.target

# Separando treino e teste
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.3, random_state=42)

# Criando o modelo
arvore = DecisionTreeClassifier()

# Treinando
arvore.fit(X_treino, y_treino)

# Plotando a árvore
plt.figure(figsize=(20,10))
plot_tree(arvore, filled=True, rounded=True, class_names=cancer.target_names, feature_names=cancer.feature_names)
plt.title("Árvore de Decisão - Câncer de Mama")
plt.show()

# Fazendo previsões
r = arvore.predict(X_teste)

# Calculando acurácia
acuracia = accuracy_score(y_teste, r)
print("Acurácia:", acuracia)
```
